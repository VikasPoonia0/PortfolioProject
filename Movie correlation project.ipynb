```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots we will create



# read in the data
df = pd.read_csv(r'C:\Users\Vikas\Downloads\movies.csv\movies.csv')
```

```python
#let's look at the data
df.head()
```

```python
# checking if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
#droping rows with null values
df=df.dropna()
```

```python
# data types for our columns
df.dtypes
```

```python
# changing data type of budget, gross and votes into integer

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
#create correct year column for release date
df['yearcorrect'] = df['released'].str.extract(r'(\d{4})').astype(int)
df.tail(100)
```

```python

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# drop any duplicates
df.drop_duplicates()
```

```python
# budget high correlation
# company high correlation
```

```python
# scatter plot with budget and gross


plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross', data=df, scatter_kws={"color":"orange"}, line_kws={"color":"blue"})
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```
